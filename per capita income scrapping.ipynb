# Basic EDA


```python
df = pd.read_csv("per_capita_income_state_wise.csv",header=None)
```

```python
df.columns = df.iloc[0]
```

```python
df = df[1:]
```

```python
df
```

```python
df.head()
```

```python
# check all columns of the dataset
df.columns
```

```python
# check information of the dataset
df.info()
```

```python
# check descreption of the dataset
df.describe(include = "all")
```

```python
# checking null values
df.isnull().sum()
```

```python
# dropping columns that i won't need
df.drop(['Per Capita Income ($) 2020-21 (based on current price)'],axis = 1,inplace = True)
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
# Create a bar chart
plt.figure(figsize=(20, 10))
sns.set_theme(font_scale=1.5)
plt.bar(df['State/ UT'], df['Per Capita Income (INR) 2020-21 (based on current price)'], color='skyblue')
plt.xlabel('State/ UT')
plt.ylabel('Per Capita Income (INR) 2020-21 (based on current price)')
plt.title('Per Capita Income of Indian States')
plt.xticks(rotation=90)
plt.tight_layout()

# Display the chart
plt.show()
```

from this graph it shows that goa is the highest per capita income state and bihar is the lowest.
